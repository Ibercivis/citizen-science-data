# Restaurar un notebook al original

Tus cambios viven solo en tu copia del Hub; el original siempre está en [GitHub](https://github.com/Ibercivis/citizen-science-data). Escribe abajo el nombre del notebook que quieres restaurar (sin `.ipynb`), o `*` para restaurarlos **todos**, y ejecuta la celda.

⚠️ Restaurar descarta tus cambios en ese fichero. Si quieres conservarlos, renombra antes tu copia (clic derecho → *Rename*).

```python
NOTEBOOK = ''   # ← nombre sin .ipynb, o '*' para todos

from pathlib import Path
import subprocess

REPO = Path.home() / 'citizen-science-data'
assert NOTEBOOK, "Escribe el nombre en la variable NOTEBOOK y vuelve a ejecutar."
objetivo = '.' if NOTEBOOK == '*' else f'{NOTEBOOK}.ipynb'
subprocess.run(['git', 'fetch', '-q', 'origin'], cwd=REPO, check=True)
subprocess.run(['git', 'checkout', 'origin/main', '--', objetivo], cwd=REPO, check=True)
print(f"✔ Restaurado: {objetivo}")
print("Si lo tienes abierto: File → Reload Notebook from Disk.")
```
