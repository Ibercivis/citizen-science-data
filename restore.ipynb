# Restore a notebook to the original

Your changes only live in your Hub copy; the original is always on [GitHub](https://github.com/Ibercivis/citizen-science-data). Type below the name of the notebook you want to restore (without `.ipynb`), or `*` to restore them **all**, and run the cell.

⚠️ Restoring discards your changes to that file. To keep them, rename your copy first (right-click → *Rename*).

```python
NOTEBOOK = ''   # ← name without .ipynb, or '*' for all

from pathlib import Path
import subprocess

REPO = Path.home() / 'citizen-science-data'
assert NOTEBOOK, "Type the name in the NOTEBOOK variable and run again."
target = '.' if NOTEBOOK == '*' else f'{NOTEBOOK}.ipynb'
subprocess.run(['git', 'fetch', '-q', 'origin'], cwd=REPO, check=True)
subprocess.run(['git', 'checkout', 'origin/main', '--', target], cwd=REPO, check=True)
print(f"✔ Restored: {target}")
print("If it's open: File → Reload Notebook from Disk.")
```
